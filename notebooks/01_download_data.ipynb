```python
import pandas as pd
import yfinance as yf
from pathlib import Path
```

```python
# Create directory if it doesn't exist
Path("../data/raw").mkdir(parents=True, exist_ok=True)
```

```python
# Download S&P 500 historical data
sp500 = yf.download(
    "^GSPC",
    start="2000-01-01",
    end="2026-01-01",
    interval="1d",
    auto_adjust=True,
    progress=False,
)
```

```python
sp500 = sp500[["Close"]]

sp500.head()
```

```python
print(sp500.info())

print()

print(sp500.describe())
```

```python
sp500.to_csv("../data/raw/sp500.csv")

print("Saved to data/raw/sp500.csv")
```
